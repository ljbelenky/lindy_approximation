```python
p = 1/40
K = 50
k = np.linspace(1,K,K)
```

```python
# Timeout = (1/K)*(1-(k-1)/K)*(1-p)**(k-1)
Timeout = (1-p)**(k)*(1/(K-k+1))*(1-(k-1)/K)
```

```python
# Succeed = ((1-p)**(k-1))*(1-(k/K))*p
Succeed = p*(1-p)**(k-1)*(1-(k-1)/K)
```

```python
N = 100000
df = pd.DataFrame()
df['true_roll'] = geom(p).rvs(N)
df['timeout'] = np.random.randint(1, K+1, size = N)
df['observation'] = np.where(df['true_roll']<=df['timeout'], df['true_roll'], np.nan)
df['success'] = ~df['observation'].isna()
df['last_roll'] = np.minimum(df['timeout'], df['true_roll'])
df
```

```python
df['success'].mean()
```

```python
z = df.groupby(['success','last_roll']).count()['true_roll'].rename({'true_roll':'count'})

(z.unstack(level=0)/N).plot()
plt.plot(k, Timeout)
plt.plot(k, Succeed);
```

```python
plt.plot(k, Timeout)
plt.plot(k, Succeed);
```

```python
plt.plot(k, Succeed/(Timeout+Succeed));
```

```python
SR = df.groupby('last_roll')['success'].mean()
SR.plot()
plt.plot(k, Succeed/(Timeout+Succeed));
```

```python
N_obs = df.groupby('last_roll')['timeout'].count()/len(df)
plt.plot(N_obs);
```

```python
# plt.plot(1-(df['last_roll'].max()-k+1)/(SR-1));
K_observed = df['last_roll'].max()
p_hat = SR/(k*(SR-1)-K_observed*SR+K_observed+1)
plt.plot(p_hat)
plt.axhline((p_hat*N_obs).sum(), color = 'red')
plt.title((p_hat*N_obs).sum());
```
